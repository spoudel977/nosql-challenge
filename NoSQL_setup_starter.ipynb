2. Find the BusinessTypeID for "Restaurant/Cafe/Canteen" and return only the `BusinessTypeID` and `BusinessType` fields.

```python
# Find the BusinessTypeID for "Restaurant/Cafe/Canteen" and return only the BusinessTypeID and BusinessType fields

restaurant_type_query = {"BusinessType": "Restaurant/Cafe/Canteen"}
projection = {"BusinessTypeID": 1, "BusinessType": 1, "_id": 0}

restaurant_type_result = establishments.find(restaurant_type_query, projection)

for restaurant in restaurant_type_result:
    pprint(restaurant)
```

3. Update the new restaurant with the `BusinessTypeID` you found.

```python
# Update the new restaurant with the correct BusinessTypeID

# Find the BusinessTypeID for "Restaurant/Cafe/Canteen"
restaurant_type_query = {"BusinessType": "Restaurant/Cafe/Canteen"}
projection = {"BusinessTypeID": 1, "_id": 0}

restaurant_type_result = establishments.find_one(restaurant_type_query, projection)

# Update the new restaurant with the correct BusinessTypeID
if restaurant_type_result:
    business_type_id = restaurant_type_result.get("BusinessTypeID")
    update_query = {"BusinessName": "Penang Flavours"}
    update_values = {"$set": {"BusinessTypeID": business_type_id}}

    update_result = establishments.update_one(update_query, update_values)

    if update_result.modified_count > 0:
        print("New restaurant updated with BusinessTypeID:", business_type_id)
    else:
        print("New restaurant not updated. BusinessTypeID not found.")
else:
    print("BusinessTypeID for 'Restaurant/Cafe/Canteen' not found.")
```

```python
# Confirm that the new restaurant was updated
# Query the updated restaurant document to confirm the BusinessTypeID update
updated_restaurant = establishments.find_one({"BusinessName": "Penang Flavours"})

if updated_restaurant:
    print("Updated restaurant details:")
    pprint(updated_restaurant)
else:
    print("New restaurant not found. Update may not have been successful.")

```

5. Some of the number values are stored as strings, when they should be stored as numbers.

Use `update_many` to convert `latitude` and `longitude` to decimal numbers.

```python
from decimal import Decimal

# Update the data type from String to Decimal for longitude and latitude
update_result = establishments.update_many({}, [
    {"$set": {"geocode.longitude": float(Decimal("0.0"))}},
    {"$set": {"geocode.latitude": float(Decimal("0.0"))}}
])

# Check the number of documents updated
updated_count = update_result.modified_count
print("Number of documents updated with Decimal longitude and latitude:", updated_count)
```

Use `update_many` to convert `RatingValue` to integer numbers.

```python
# Set non 1-5 Rating Values to Null
non_ratings = ["AwaitingInspection", "Awaiting Inspection", "AwaitingPublication", "Pass", "Exempt"]
establishments.update_many({"RatingValue": {"$in": non_ratings}}, [ {'$set':{ "RatingValue" : None}} ])
```

```python
# Change the data type from String to Integer for RatingValue

update_result = establishments.update_many({}, [
    {"$set": {"RatingValue": {"$toInt": "$RatingValue"}}}
])

# Check the number of documents updated
updated_count = update_result.modified_count
print("Number of documents updated with integer RatingValue:", updated_count)

```

```python
# Check that the coordinates and rating value are now numbers
# Find one document to check the data types of coordinates and RatingValue
sample_document = establishments.find_one()

if sample_document:
    longitude = sample_document.get("geocode", {}).get("longitude")
    latitude = sample_document.get("geocode", {}).get("latitude")
    rating_value = sample_document.get("RatingValue")

    if isinstance(longitude, (float, int)) and isinstance(latitude, (float, int)) and isinstance(rating_value, int):
        print("Coordinates (longitude and latitude) and RatingValue are now numbers.")
    else:
        print("Coordinates (longitude and latitude) and/or RatingValue are not numbers.")
else:
    print("No document found in the collection.")

```
